```python
# What this cell does: checks that all libraries installed correctly
# Run this FIRST before anything else

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("pandas version  :", pd.__version__)
print("numpy version   :", np.__version__)
print("All libraries loaded successfully ")
```

```python
# What this cell does: tells Python where to find your data files
# Using relative paths so the project works on any computer

import os

# Define paths
RAW_DATA_PATH = os.path.join('..', 'data', 'raw')
PROCESSED_DATA_PATH = os.path.join('..', 'data', 'processed')

print("Raw data folder      :", RAW_DATA_PATH)
print("Processed data folder:", PROCESSED_DATA_PATH)

# Check if the files actually exist
for file in ['POSOCO_data.csv', 'CEA_DGR_data.csv']:
    full_path = os.path.join(RAW_DATA_PATH, file)
    status = " Found" if os.path.exists(full_path) else " NOT FOUND"
    print(f"{file}: {status}")
```

```python
# BEFORE: We have not looked at the data at all
# Operation: Load raw POSOCO data and inspect it
# WHY: You must understand the data before touching it

posoco_raw = pd.read_csv(os.path.join(RAW_DATA_PATH, 'POSOCO_data.csv'))

print("=== BEFORE: Raw POSOCO Data ===")
print(f"Shape: {posoco_raw.shape}")
print(f"\nColumn names:\n{posoco_raw.columns.tolist()}")
print(f"\nData types:\n{posoco_raw.dtypes}")
print(f"\nFirst 5 rows:")
posoco_raw.head()
```

```python
# Checking all column names clearly
for i, col in enumerate(posoco_raw.columns.tolist()):
    print(i, col)
```

```python
# BEFORE: We don't know how dirty the data is
# WHY: Nulls and wrong dtypes break every model downstream

print("=== NULL VALUES (top 20 columns) ===")
print(posoco_raw.isnull().sum().head(20))

print("\n=== DATA TYPE OF DATE COLUMN ===")
print(type(posoco_raw['yyyymmdd'][0]), posoco_raw['yyyymmdd'][0])

print("\n=== BASIC STATS (first 8 columns) ===")
posoco_raw.iloc[:, :8].describe()
```

```python
# WHY: We need to know what % of data is missing
# before deciding whether to drop or fill nulls

total_rows = len(posoco_raw)
print(f"Total rows: {total_rows}")

print("\n=== NULL % FOR KEY COLUMNS ===")
null_pct = (posoco_raw.isnull().sum() / total_rows * 100).round(2)
print(null_pct[null_pct > 0].to_string())
```

```python
# Final BEFORE snapshot before any cleaning
print("=== BEFORE CLEANING SUMMARY ===")
print(f"Shape                : {posoco_raw.shape}")
print(f"Date dtype           : {posoco_raw['yyyymmdd'].dtype}")
print(f"Duplicate rows       : {posoco_raw.duplicated().sum()}")
print(f"Total missing values : {posoco_raw.isnull().sum().sum()}")
print(f"Date range           : {posoco_raw['yyyymmdd'].min()} to {posoco_raw['yyyymmdd'].max()}")
```
